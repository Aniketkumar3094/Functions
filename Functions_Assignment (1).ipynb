Q 6. Explain the concept of generators in Python and how they are defined.

Ans--In Python, generators are a special type of iterator that allow you to produce a sequence of values lazily — meaning values are generated on the fly and not stored in memory all at once.

They are useful for working with large datasets or infinite sequences where storing all values in memory would be inefficient or impossible.

**Key Points About Generators**

1. Lazy Evaluation

   - Values are produced only whenrequested (via iteration).
   - Saves memory and improves performance.
2. Iterator Protocol

   - Generators automatically implement the iterator protocol (__iter__() and __next__()).
3. One-time Use

   - Once a generator is exhausted, it cannot be reused; you must create it again.
**Defining Generators**

There are two main ways to define generators in Python:

1. **Generator Functions**

   - Defined like a normal function but use the yield keyword instead of return.
   - Each yield pauses the function, saving its state, and resumes from there on the next iteration.



```python
def count_up_to(n):
    """Generator that yields numbers from 1 to n."""
    num = 1
    while num <= n:
        yield num  # Pause and return the current value
        num += 1

# Using the generator
for number in count_up_to(5):
    print(number)
```

2. **Generator Expressions**

   - Similar to list comprehensions but use parentheses () instead of square brackets [].
   - More concise for simple generators.


```python
squares = (x * x for x in range(5))
for sq in squares:
    print(sq)
```

Q 10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans--In Python, map(), filter(), and reduce() are functional programming tools that operate on iterables (like lists, tuples, etc.) but serve different purposes:

1. **map(function, iterable, ...)**

   - **Purpose:** Applies a given function to each element of one or more iterables and     returns a map object (iterator) with the results.
   - **Key Point:** Transforms every element.

2. **filter(function, iterable)**

   - **Purpose:** Filters elements from the iterable for which the function returns True.
   - **Key Point:** Selects a subset of elements.
3. **reduce(function, iterable[, initializer])**

   - **Purpose:**  Applies a function cumulatively to the elements of the iterable, reducing it to a single value.
   - **Key Point:**  Aggregates results.
   - **Note:**  reduce() is in the functools module, so you must import it.

Q 2.Create a Python function that accepts a string and returns the reverse of that string.

```python
def reverse_string(s):
    # Validate input type
    if not isinstance(s, str):
        raise TypeError("Input must be a string.")

    # Reverse using slicing (fast and Pythonic)
    return s[::-1]


# Example usage:
try:
    text = "Hello World"
    print("Original:", text)
    print("Reversed:", reverse_string(text))
except Exception as e:
    print("Error:", e)
```

Q 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

```python
def square_numbers(numbers):

    if not isinstance(numbers, list):
        raise TypeError("Input must be a list.")

    for num in numbers:
        if not isinstance(num, int):
            raise TypeError(f"All elements must be integers. Invalid element: {num}")

    return [num ** 2 for num in numbers]


# Example usage
if __name__ == "__main__":
    try:
        # Example input
        nums = [1, 2, 3, 4, 5]
        result = square_numbers(nums)
        print("Original list:", nums)
        print("Squared list:", result)
    except Exception as e:
        print("Error:", e)

```

Q 4.Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
import math

def is_prime(n):
    """
    Check if a number is prime.
    Returns True if prime, False otherwise.
    """
    if not isinstance(n, int) or n < 1:
        raise ValueError("Input must be a positive integer.")

    if n == 1:
        return False  # 1 is not prime
    if n == 2:
        return True   # 2 is prime
    if n % 2 == 0:
        return False  # Even numbers > 2 are not prime

    # Check divisors up to sqrt(n)
    limit = int(math.sqrt(n)) + 1
    for i in range(3, limit, 2):
        if n % i == 0:
            return False
    return True

# Test the function for numbers from 1 to 200
if __name__ == "__main__":
    primes = [num for num in range(1, 201) if is_prime(num)]
    print("Prime numbers from 1 to 200 are:")
    print(primes)
```

Q 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    """Iterator class to generate Fibonacci sequence up to n terms."""

    def __init__(self, n_terms):
        # Validate input
        if not isinstance(n_terms, int) or n_terms < 0:
            raise ValueError("Number of terms must be a non-negative integer.")

        self.n_terms = n_terms
        self.count = 0
        self.a, self.b = 0, 1  # First two Fibonacci numbers

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b


# Example usage
if __name__ == "__main__":
    try:
        n = int(input("Enter the number of Fibonacci terms: "))
        fib_iter = FibonacciIterator(n)

        print(f"Fibonacci sequence up to {n} terms:")
        for num in fib_iter:
            print(num, end=" ")
        print()

    except ValueError as e:
        print(f"Error: {e}")
```

Q 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    """
    Generator that yields powers of 2 from 2^0 up to 2^max_exponent.

    Args:
        max_exponent (int): The maximum exponent (must be >= 0).

    Yields:
        int: The next power of 2 in the sequence.
    """
    if not isinstance(max_exponent, int):
        raise TypeError("max_exponent must be an integer.")
    if max_exponent < 0:
        raise ValueError("max_exponent must be non-negative.")

    for exp in range(max_exponent + 1):
        yield 2 ** exp


# Example usage
if __name__ == "__main__":
    try:
        n = int(input("Enter the maximum exponent (>= 0): "))
        print(f"Powers of 2 up to 2^{n}:")
        for value in powers_of_two(n):
            print(value)
    except (ValueError, TypeError) as e:
        print(f"Error: {e}")
```

Q 7.. Implement a generator function that reads a file line by line and yields each line as a string

```python
def read_file_line_by_line(file_path):
   """
   A generator function to read a file line by line

   Args:
       file_path (str): Path to the file to be read

   Yields:
       str: The next line in the file
   """
   with open(file_path, 'r') as file:
       for line in file:
           yield line.strip() # Remove trailing newline character
```

Q 9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
# Python program to convert a list of Celsius temperatures to Fahrenheit using map()

def celsius_to_fahrenheit(celsius):
    """
    Convert Celsius to Fahrenheit using the formula:
    F = (C * 9/5) + 32
    """
    return round((celsius * 9 / 5) + 32, 2)  # Rounded to 2 decimal places

def main():
    try:
        # Take input from the user as a comma-separated list
        user_input = input("Enter temperatures in Celsius (comma-separated): ").strip()

        if not user_input:
            print("No input provided.")
            return

        # Convert input string to a list of floats
        celsius_list = list(map(float, user_input.split(",")))

        # Use map() to convert each Celsius value to Fahrenheit
        fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))

        # Display results
        print("\nCelsius temperatures:   ", celsius_list)
        print("Fahrenheit temperatures:", fahrenheit_list)

    except ValueError:
        print("Invalid input. Please enter only numeric values separated by commas.")

if __name__ == "__main__":
    main()
```

Q 10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
# Python program to remove all vowels from a given string using filter()

def remove_vowels(input_str):
    """
    Removes all vowels from the given string using filter().
    Handles both uppercase and lowercase vowels.
    """
    vowels = "aeiouAEIOU"
    # filter() keeps only characters that are NOT vowels
    return ''.join(filter(lambda ch: ch not in vowels, input_str))

def main():
    try:
        # Take input from the user
        user_input = input("Enter a string: ").strip()

        # Validate input (must be non-empty)
        if not user_input:
            print("Error: Empty string provided.")
            return

        # Remove vowels
        result = remove_vowels(user_input)

        # Display result
        print("String without vowels:", result)

    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```
